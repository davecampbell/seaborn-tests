```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.DataFrame(np.random.randint(0,100, size=(100,4)), columns=list('ABCD'))
```

```python
df.head()
```

```python
df['TENS-A'] = df.apply(lambda x: str(int(x.A / 10)), axis=1)
```

```python
df.head()
```

```python
sns.distplot(df.A)
```

```python
grouped = df.groupby('TENS-A')
```

```python
# one plot for all groups
for name, group in grouped:
    sns.distplot(group.A)
```

```python
# separate plot for each group
ncols=1
nrows = int(np.ceil(grouped.ngroups/ncols))

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12,8), sharey=True, sharex=True)

for (key, ax) in zip(grouped.groups.keys(), axes.flatten()):
    grouped.get_group(key).plot(ax=ax)

# for name, group in grouped:
#    sns.distplot(group.A)
```
